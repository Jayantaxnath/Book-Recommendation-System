### 1. Popularity Based Recommender System

```python
rating_wbooks = ratings.merge(books, on='ISBN')
```

```python
print(rating_wbooks.shape)
rating_wbooks.head()
```

```python
num_rating_df = rating_wbooks.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = (
    rating_wbooks.groupby('Book-Title')['Book-Rating']
    .mean()
    .reset_index()
    .rename(columns={'Book-Rating': 'avg_ratings'})
)
avg_rating_df.head()
```

```python
num_rating_df['avg_ratings'] = avg_rating_df['avg_ratings'].copy()
num_rating_df
```

```python
popular_df = num_rating_df[num_rating_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
top_50_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
print(top_50_df.shape)
top_50_df.head(3)
```

```python
top_50_df.columns
```

```python
top_50_df_final = top_50_df[['Book-Title','Book-Author', 'num_ratings', 'avg_ratings','Image-URL-M']]
```

### 2. Collaborative Based Recommender System

```python
rating_with_books = rating_wbooks.copy()
```

```python
rating_with_books.head()
```

```python
x = rating_with_books.groupby('User-ID').count()['Book-Rating'] > 200
rating_more_than_200 = x[x].index
rating_more_than_200
```

```python
filtered_rating = rating_with_books[rating_with_books['User-ID'].isin(rating_more_than_200)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
famous_book
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings.head()
```

```python
final_ratings.columns
```

```python
pivot = final_ratings.pivot_table(index=['Book-Title'], columns='User-ID', values='Book-Rating')
```

```python
pivot
```

```python
pivot.fillna(0, inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pivot)
```

```python
def book_recommendation(book_name):
    #index fetch
    index = np.where(pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x: x[1], reverse=True)[1:6]
    recommended_books = []
    for item in (similar_items):
        data = []
        temp_df = books[books['Book-Title'] == pivot.index[item[0]]]
        data.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        data.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        data.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        recommended_books.append(data)
    return recommended_books
```

```python
# rb = book_recommendation('The Da Vinci Code')
# rb = book_recommendation('Angels &amp; Demons')
rb = book_recommendation('1984')
```
